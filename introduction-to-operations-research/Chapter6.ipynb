### Problem 6.1.1

(a.1)

* Given the following primal problem:
$$
\max \quad Z = 2x_1 + x_2 \\
s.t.
\begin{cases}
x_2 ≤ 10 \\
2x_1 + 5x_2 ≤ 60 \\
x_1 + x_2 ≤ 18 \\
3x_1 + x_2 ≤ 44 \\
x_1, x_2 \geq 0
\end{cases}
$$

* We can express this problem in the standard form as follows:
$$
\max \quad Z = 2x_1 + x_2 + 0w_1 + 0w_2 + 0w_3 + 0w_4 \\
s.t.
\begin{cases}
x_2 + w_1 = 10 \\
2x_1 + 5x_2 + w_2 = 60 \\
x_1 + x_2 + w_3 = 18 \\
3x_1 + x_2 + w_4 = 44 \\
x_1, x_2, w_1, w_2, w_3, w_4 \geq 0
\end{cases}
$$

* Hence, we have the following matrices: $\mathbf{c} = \begin{bmatrix} 2 \\ 1 \\ 0 \\ 0 \\ 0 \\ 0 \end{bmatrix}$, 
$\mathbf{A} = \begin{bmatrix} 0 & 1 & 1 & 0 & 0 & 0 \\ 2 & 5 & 0 & 1 & 0 & 0 \\ 1 & 1 & 0 & 0 & 1 & 0 \\ 3 & 1 & 0 & 0 & 0 & 1 \end{bmatrix}$, and $\mathbf{b} = \begin{bmatrix} 10 \\ 60 \\ 18 \\ 44 \end{bmatrix}$.

* Therefore, the corresponding dual problem is given as:
$$
\min \quad 10y_1 + 60y_2 + 18y_3 + 44y_4 \\
\begin{cases}
0y_1 + 2y_2 + y_3 + 3y_4 \geq 2 \\
y_1 + 5y_2 + y_3 + y_4 \geq 1 \\
y_1, y_2, y_3, y_4 \geq 0
\end{cases}
$$

(a.2)

$$

x_1 + x_2 ≤ 12 \\
5x_1 + 3x_2 ≤ 45 \\
x_1, x_2 \geq 0

$$

$$
\max \quad Z = 10x_1 + 20x_2 + 0w_1 + 0w_2 + 0w_3 \\

\begin{cases}
-x_1 + 2x_2 + w_1 = 15 \\
x_1 + x_2 + w_2 =  12 \\
5x_1 + 3x_2 + w_3 = 45 \\
x_1, x_2, w_1, w_2, w_3 \geq 0
\end{cases}
$$

* Hence, we have the following matrices: $\mathbf{c} = \begin{bmatrix} 10 \\ 20 \\ 0 \\ 0 \\ 0 \end{bmatrix}$, 
$\mathbf{A} = \begin{bmatrix} -1 & 2 & 1 & 0 & 0 \\ 1 & 1 & 0 & 1 & 0 \\ 5 & 3 & 0 & 0 & 1 \end{bmatrix}$, and $\mathbf{b} = \begin{bmatrix} 15 \\ 12 \\ 45 \end{bmatrix}$.

* Therefore, the corresponding dual problem is given as:
$$
\min \quad 15y_1 + 12y_2 + 45y_3 \\
\begin{cases}
-y_1 + y_2 + 5y_3 \geq 10 \\
2y_1 + y_2 + 3y_3 \geq 20 \\
y_1, y_2, y_3, y_4 \geq 0
\end{cases}
$$

(b.1)

* Given the following primal problem:
$$
\max \quad Z = 2x_1 + 3x_2 \\
s.t.
\begin{cases}
−3x_1 + x_2 ≤ 1 \\
−4x_1 + 2x_2 ≤ 20 \\
−4x_1 − x_2 ≤ 10 \\
−x_1 + 2x_2 ≤ 5 \\
x_1 ≥ 0, x_2 ≥ 0
\end{cases}
$$

* We can express this problem in the standard form as follows:
$$
\max \quad Z = 2x_1 + 3x_2 \\
s.t.
\begin{cases}
−3x_1 + x_2 + w_1 = 1 \\
−4x_1 + 2x_2 + w_2 = 20 \\
−4x_1 − x_2 + w_3 = 10 \\
−x_1 + 2x_2 + w_4 = 5 \\
x_1 ≥ 0, x_2 ≥ 0
\end{cases}
$$

* Hence, we have the following matrices: $\mathbf{c} = \begin{bmatrix} 2 \\ 3 \\ 0 \\ 0 \\ 0 \\ 0 \end{bmatrix}$, 
$\mathbf{A} = \begin{bmatrix} -3 & 1 & 1 & 0 & 0 & 0 \\ -4 & 2 & 0 & 1 & 0 & 0 \\ −4 & −1 & 0 & 0 & 1 & 0 \\ −1 & 2 & 0 & 0 & 0 & 1 \end{bmatrix}$, 
and $\mathbf{b} = \begin{bmatrix} 1 \\ 20 \\ 10 \\ 15 \end{bmatrix}$.

* Therefore, the corresponding dual problem is given as:
$$
\min \quad Z = y_1 + 20y_2 + 10y_3 + 5y_4 \\
s.t.
\begin{cases}
−3y_1 - 4y_2 -4y_3 - y_4 \geq 2 \\
y_1 + 2y_2 - y_3 + 2y_4 \geq 3 \\
y_1, y_2, y_3, y_4 \geq 0
\end{cases}
$$

3x_1 + 4x_2 + w_2 = 60 \\

x_1 ≥ 0; x_2 ≥ 0; w_1, w_2, w_3 \geq 0 \\
\end{cases}

$$
\max \quad Z = 42y_1 + 60y_2 + 18y_3 \\

2y_1 + 3y_2 + y_3 \geq 5 \\
